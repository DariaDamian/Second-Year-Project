# Testing the Google embeddings

```
df = pd.DataFrame(read_processed_data('/work/nlp-project/data/processed/train.conll'))
train_vocab = np.unique(df[0].explode())
```

```
unknowns = []
for word in train_vocab:
    if not GoogleEmbs.__contains__(word):
        unknowns.append(word)
```

```
print(len(train_vocab), len(unknowns))
unknowns
```

The google model has no embeddings for a bunch tokens: symbols, names, abbreviations, urls, misspellings, punctuation.

What words are near the zero vector in the embedding space?

```
GoogleEmbs.most_similar(positive=np.zeros(300), topn=50)
```
